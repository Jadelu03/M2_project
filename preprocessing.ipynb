```python
file_path = 'E:\Year_2_Semester_1\chemometrics\Exercises\M2\M2_project\M2_SENSORY_format.csv'
df = pd.read_csv(file_path)
df
```

There is one blank cell of sample 18 at column 'Melting"

```python
def fill_blank_values(df, column, group_cols, method='median'):
  groups = df.groupby(group_cols)

  for name, group in groups:
    if method == 'mean':
      fill_value = group[column].mean()
    elif method == 'median':
      fill_value = group[column].median()
    else:
      print(f"method must be either 'mean' or 'median'")

    missing_count = group[column].isna().sum()
    if missing_count >0:
      print(f"Filling {missing_count} missing values in group {name} with {method}: {fill_value}")
      df.loc[group.index, column] = df.loc[group.index, column].fillna(fill_value)

  return df

group_columns = ['FiberType', 'Pressure','Concentrate']
data_filled_median = fill_blank_values(df,'Melting', group_columns, method='median')
data_filled_mean = fill_blank_values(df,'Melting', group_columns, method='mean')
data_filled_median.to_csv('filled_medain.csv', index=False)
data_filled_mean.to_csv('filled_mean.csv', index=False)
    

```

By checking the two csv tables, the mean and median values are both 7.55
